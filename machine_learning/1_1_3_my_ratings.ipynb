```python
import pandas as pd
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/ml-latest-small/ratings.csv')
ratings.sample(5)
```

### timestamp 컬럼 처리

```python
# datetime 모듈에서 datetime 클래스를 현재 스코프로 가져옴
from datetime import datetime
```

```python
# datetime.fromtimestamp(x)를 사용하여 Unix 타임스탬프를 일반적인 날짜와 시간 형식으로 변환
# datetime.fromtimestamp(990062737)
# lambda 함수로 전체 적용

ratings['timestamp'] = ratings['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
ratings.sample()
```

### 내 데이터 붙이기

```python
myratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/ml-latest-small/my_ratings.csv')
myratings.sample()
```

```python
# dtype 확인
myratings['timestamp'].sample()

# 해석
# timestamp가 문자열(object)이기 때문에 날짜(datetime)로 바꿔줘야 한다.
```

```python
# 날짜타입으로 변환
# 덮어쓰기 필수!

myratings['timestamp'] = pd.to_datetime(myratings['timestamp'])
myratings['timestamp'].sample()
```

```python
# 데이터 이어붙이기

ratings_cancat = pd.concat([ratings, myratings])
ratings_cancat.tail(5)
```

```python
# 피클로 저장하기
ratings_cancat.to_pickle('/content/drive/MyDrive/Colab Notebooks/data/ml-latest-small/ratings_updated.p')
```

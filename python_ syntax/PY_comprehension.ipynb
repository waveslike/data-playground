# Q1.

리스트 컴프리헨션을 이용해서 1~10까지의  홀수를 구하세요.


💡참고

```python
>>> sample = [i for i in range(1, 9) if i % 2 == 0]
>>> print(sample)

```

````python
# Answer

odd = [i for i in range(1, 11) if i % 2 == 1]

print(odd)
````


# Q2.

`enumerate` 내장 함수를 사용해서 my_pokemon 리스트의 인덱스와 원소를 출력하세요.

```
enumerate(iterable, start=0)
```
enumerate() 함수는 파이썬에서 사용되는 내장 함수 중 하나로, 순회 가능한(iterable) 객체(리스트, 튜플, 문자열 등)를 입력으로 받아 인덱스와 해당 요소를 포함하는 튜플을 순차적으로 생성하는 iterator를 반환합니다.


```
fruits = ['apple', 'banana', 'cherry']

for index, value in enumerate(fruits):
    print(f'Index: {index}, Value: {value}')

Index: 0, Value: apple
Index: 1, Value: banana
Index: 2, Value: cherry
```




💡참고.

```python
>>> number = [10, 20, 30]
>>> for i, number in enumerate(number):
>>> print(i, number)

0 10
1 20
2 30
```

````python
my_pokemon = ['피카츄', '꼬북이', '파이리', '이상해씨']
````

````python
# Answer 1

for i, my_pokemon in enumerate(my_pokemon):
    print(i, my_pokemon)
````

````python
my_pokemon = ['피카츄', '꼬북이', '파이리', '이상해씨']
````

````python
# Answer 12
for i, my_pokemon in enumerate(my_pokemon):
    print(f'Index: {i}, Value: {my_pokemon}')
````

# Q3.



`zip` 함수를 이용하여 fruit 리스트와 color 리스트를 묶고, 아래와 같은 **리스트**를 만들어주세요.

```python
[('apple', 'red'), ('banana', 'yellow'), ('orange', 'orange')]
```

💡참고.

```python
>>> pokemon = ['피카츄', '꼬북이']
>>> hp = [100, 105]
>>> pokemon_hp = list(zip(pokemon, hp))
>>> print(pokemon_hp)

[('피카츄', 100), ('꼬북이', 105)]
```

````python
fruit = ['apple', 'banana', 'orange']
color = ['red', 'yello', 'orange']
````

````python
# Answer

fruit_color = list(zip(fruit, color))

print(fruit_color)
````

# Q4.

`zip` 함수를 이용하여 fruit 리스트와 color 리스트를 묶고, dict 컴프리헨션을 이용하여 아래와 같은 **딕셔너리**를 만들어주세요.

```python
{'apple': 'red', 'banana': 'yello', 'orange': 'orange'}
```

💡참고.

```python
>>> pokemon = ['피카츄', '꼬북이']
>>> hp = [100, 105]
>>> pokemon_hp = {k:v for k, v in zip(pokemon, hp)}
>>> print(pokemon_hp)

{'피카츄': 100, '꼬북이': 105}
```

````python
fruit = ['apple', 'banana', 'orange']
color = ['red', 'yello', 'orange']
````

````python
# Answer

{k:v for k, v in zip(fruit, color)}
````

# Q5.

`dict 컴프리헨션`을 이용하여 아래와 같이 key, vaule 값을 바꿔 **딕셔너리**를 출력하세요.

```python
{'red': 'apple', : 'yello' : 'banana', 'orange': 'orange'}
```

💡참고.

```python
>>> pokemon = ['피카츄', '꼬북이']
>>> hp = [100, 105]
>>> key_change = {h: n for n, h in zip(pokemon, hp)}
>>> print(key_change)

{100: '피카츄', 105: '꼬북이'}
```

````python
fruit = ['apple', 'banana', 'orange']
color = ['red', 'yello', 'orange']
````

````python
# Answer

{k: v for v, k in zip(fruit, color)}
````

# Q6.

`dict 컴프리헨션`을 이용하여 아래와 같이 key, vaule 값을 바꿔 **딕셔너리**를 출력하세요.
```python
{'red': 'apple', : 'yello' : 'banana', 'orange': 'orange'}
```

💡참고.

```python
>>> pokemon = {'피카츄': 100, '꼬북이': 105}
>>> key_change = {h: n for n, h in pokemon.items()}
>>> print(key_change)

{100: '피카츄', 105: '꼬북이'}
```

`items()` 메서드를 호출하면 딕셔너리의 각 항목을 (키, 값)의 튜플 형태로 가지는 dict_items 뷰 객체가 반환됩니다.

````python
fruit_color = {'apple': 'red', 'banana': 'yello', 'orange': 'orange'}
````

````python
# Answer

{k: v for v, k in fruit_color.items()}
````

# Q7.

`try`와 `except`을 이용하여 `1/0` 코드를 실행했을 때 발생하는 오류 메시지를 출력해주세요.

💡참고.
```python
>>> try:
>>>     '1' + 0
>>> except TypeError as e:
>>>     print(f'에러메시지: {e}')

에러메시지: can only concatenate str (not "int") to str
```


````python
# Answer

try:
    1 / 0
except ZeroDivisionError as e:
    print(f'에러메시지: {e}')
````

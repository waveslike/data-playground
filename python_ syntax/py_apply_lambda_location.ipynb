### Python lambda

```python
# def == define, definition
def add_1(x):
    return x + 1
```

```python
add_1(1)
```

```python
# lambda
add_1 = lambda x: x + 1
```

```python
add_1(1)
```

### 정규표현식을 사용하지 않고, 개봉연도 추출하기

```python
import pandas as pd
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/ml-latest-small/movies.csv', index_col='movieId')
movies.sample()
```

```python
# 타이틀 1번째
sample_title = movies['title'][1]
sample_title
```

```python
sample_title[:3] # 0~2 index
```

```python
sample_title[-5:-1]
```

```python
# 함수로 만들면
def get_year(x):
    return x[-5:-1]
```

```python
get_year(sample_title)
```

```python
# 방법0: for문
years = []
for title in movies['title']:
    years.append(get_year(title))
```

```python
# 중복값 제거
print(set(years))
```

```python
# 방법1: python apply와 lambda lambda expression 사용하여 간단하게 풀어보기

# movies['title'].apply(get_year)
movies['year'] = movies['title'].apply(lambda x: x[-5:-1])
movies['year'].unique()


# 설명
# apply: title이라는 컬럼에 함수를 반복 적용, for 문의 기능.
# apply(함수): 함수를 lambda 함수로도 써 줄 수 있다.

# 해석
# '012)', '973)'와 같은 정제되지 않은 값들을 알아보자.
```

```python
movies[movies['year'] == '012)']
```

```python
# 정제되지 않은 셀 직접 출력해보기
# pandas .loc[] == location
# movies.loc[index, column]

movies.loc[102084, 'title']

# 해석
# 'Justice League: Doom (2012) ' << 공백 한칸 있음
# 따라서 'title' 앞뒤 공백 없애야 함
```

```python
# .strip(): 문자열의 앞뒤 공백을 없애줌
movies.loc[102084, 'title'].strip()
```

```python
# 공백 없애기 전체 적용
movies['title'] = movies['title'].apply(lambda x: x.strip())
movies.sample()
```

```python
# 연도 다시 추출
movies['year'] = movies['title'].apply(lambda x: x[-5:-1])
movies['year'].unique()

# 해석
# 앞뒤 공백 문제는 해결되었으나 문자로 출력되는 연도들이 있음
```

```python
# 남아있는 문자 errors 모아보기
errors = ['lon ', 'r On', ' Roa', 'atso', 'imal', 'erso','ligh', 'he O', 'osmo', ' Bab', 'ron ', 'irro']
```

```python
# errors 위치 확인하기
movies[movies['year'].isin(errors)]

# 해석
# errors는 개봉연도가 없는 영화들임
```

```python
# pandas .loc[] == location
# movies.loc[index, column]
# index: 필터링정보

movies.loc[movies['year'].isin(errors), 'year'] = '2050'
movies['year'].unique()

# 해석
# 'year' 컬럼 정제 완료
```

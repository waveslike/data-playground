# Q3.

값을 4개 받아서 평균을 내주는 average라는 함수를 만들어 보세요.

````python
# Answer

def average(n1, n2, n3, n4):
    result = (n1 + n2 + n3 + n4) / 4
    return result
````

# Q4.

숫자가 들어있는 리스트를 받아서 중앙값을 찾는 median라는 함수를 만들어 보세요.

<br>

#### **참고 - 평균, 중앙값, 최빈값**

평균, 중앙값, 최빈값은 숫자 데이터 집합의 중심을 나타내는 수치입니다. 이 값들은 데이터를 요약하는 '대표'하는 값으로 자주 쓰입니다.

<br>

* **평균** : 모든 수를 더하고 수의 개수로 나눠서 구함
  ```
  [4, 1, 7] 의 평균은 (4 + 1 + 7) / 3 = 12 / 3 = 4
  ```
* **중앙값** : 모든 데이터를 크기 순으로 정렬해서 가운데에 있는 데이터를 선택. (가운데에 있는 데이터가 둘이라면 두 수의 평균)
```
[4, 1, 7]를 정렬하면 (1, 4, 7) 이므로 중앙값은 가운데에 있는 4
```
```
[4, 1, 2, 7]를 정렬하면 (1, 2, 4, 7)이므로 중앙값은 (2 + 4) / 2 = 3
```
* **최빈값** : 가장 자주 나온 숫자. 데이터에서 가장 많이 등장한 숫자
```
[4, 2, 4, 3, 2, 2] 의 최빈값은 세 번 나온 2
```

````python
# 테스트 케이스
numbers1 = [1]
numbers2 = [1, 2, 3, 4, 5]
numbers3 = [1, 2, 3, 4, 5, 8]
numbers4 = [4, 1, 5, 6, 2, 6, 2, 6, 4, 2, 2, 6]
numbers5 = [1, 1, 2, 2, 2, 3, 3, 3, 3, 3, 5, 7, 7]
````

````python
# Answer

def median(numbers):
    numbers = sorted(numbers) # 리스트를 순서대로 정렬
    n = len(numbers) # 리스트의 길이

    if n % 2 == 0: # 숫자가 짝수 개 있을 때
        return (numbers[n // 2 - 1] + numbers[n // 2]) / 2 # 인덱스는 0부터 시작. numbers2 = [1, 2, 3, 4, 5] 일 때, [4] == 5.
    else: # 숫자가 홀수 개 있을 때
        return numbers[n // 2]
````

````python
median(numbers2)
````

````python
# 파이썬의 대표적인 수학 연산 라이브러리를 테스트 케이스 확인하는데 활용해봅시다.
# 테스트하고 싶은 숫자 리스트를 numbers 변수에 담아 코드를 실행한 후 값을 비교해 보세요.

# Answer

import numpy as np
np.median(numbers2)
````

# Q5.
두 수와 사칙연산자(+, -, *, /)를 받아서 두 수를 받은 순서대로 연산자로 계산해주는 calculate라는 함수를 만들어 보세요.

사칙연산자 외의 다른 값이 세번째 인자로 들어가면 ValueError를 내주세요.

```
# input 예시
1, 2, '+'
# output 예시
3
```
```
# input 예시
1, 2, '/'
# output 예시
0.5
```

[에러 메세지 출력하기 힌트](https://colab.research.google.com/drive/1kku8SByX6aaTRAZFfyXbmZuUNiV2TLy9?authuser=1#scrollTo=ZT0DnLym5E1b)

````python
# Answer

def calculate(n1, n2, operator):
    if operator == '+':
        return n1 + n2
    elif operator == '-':
        return n1 - n2
    elif operator == '*':
        return n1 * n2
    elif operator == '/':
        return n1 / n2
    else:
        raise ValueError
````

````python
calculate(1, 2, '/')
````
